```python
import pandas as pd
import numpy as np
from scipy.stats import norm, poisson, nbinom
import math
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
palette = sns.color_palette("muted")
```

```python
x_continuous = np.linspace(-10, 50, num=100)
x_discrete = np.arange(-10, 50)

# probability density function
pdf_fast_movers = norm.pdf(x_continuous, loc=20, scale=8)

# probability mass functions
pmf_medium_movers = poisson.pmf(x_discrete, mu=9)
pmf_slow_movers = nbinom.pmf(x_discrete, 2, 0.3)
```

```python
plt.figure(figsize=(6, 3))

plt.plot(x_continuous, pdf_fast_movers, label="Class A", color=palette[0], linewidth=2)
plt.plot(x_discrete, pmf_medium_movers, label="Class B", color=palette[1], linewidth=2)
plt.plot(x_discrete, pmf_slow_movers, label="Class C", color=palette[2], linewidth=2)

plt.xlim(-10, 50)

plt.title("Base Demand Distributions")
plt.xlabel("Orders", fontsize=10)
plt.ylabel("Probability", fontsize=10)

plt.legend(fontsize=10)
plt.savefig("../distribution_chart.png", bbox_inches="tight", dpi=300)
plt.show()
```
